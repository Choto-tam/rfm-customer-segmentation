```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]

df.head()
```

```python
df = df.dropna(subset=["CustomerID"])
```

```python
df["CustomerID"].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df["CustomerID"] = df["CustomerID"].astype(int)
```

```python
df = df[df["Quantity"] > 0]
```

```python
df = df[df["UnitPrice"] > 0]
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df["InvoiceDate"].min(), df["InvoiceDate"].max()
```

```python
df.info()
df.describe()
```

```python
df["CustomerID"].nunique()
```

```python
orders_per_customer = df.groupby("CustomerID")["InvoiceNo"].nunique()

orders_per_customer.head()
```

```python
revenue_per_customer = df.groupby("CustomerID")["Revenue"].sum()

revenue_per_customer.head()
```

```python
items_per_customer = df.groupby("CustomerID")["Quantity"].sum()

items_per_customer.head()
```

```python
last_purchase = df.groupby("CustomerID")["InvoiceDate"].max()

last_purchase.head()
```

```python
first_purchase = df.groupby("CustomerID")["InvoiceDate"].min()

first_purchase.head()
```

```python
customer_df = df.groupby("CustomerID").agg({
    "InvoiceNo": "nunique",
    "Revenue": "sum",
    "Quantity": "sum",
    "InvoiceDate": ["min", "max"]
})
```

```python
customer_df.columns = [
    "num_orders",
    "total_revenue",
    "total_items",
    "first_purchase",
    "last_purchase"
]
```

```python
customer_df.head()
```

```python
customer_df.shape
```

```python
analysis_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)

analysis_date
```

```python
customer_df["recency"] = (analysis_date - customer_df["last_purchase"]).dt.days
```

```python
customer_df["frequency"] = customer_df["num_orders"]
```

```python
customer_df["monetary"] = customer_df["total_revenue"]
```

```python
customer_df[["recency", "frequency", "monetary"]].head()
```

```python
customer_df["R_score"] = pd.qcut(customer_df["recency"], 5, labels=[5,4,3,2,1])
```

```python
customer_df["F_score"] = pd.qcut(customer_df["frequency"].rank(method="first"), 5, labels=[1,2,3,4,5])
```

```python
customer_df["M_score"] = pd.qcut(customer_df["monetary"], 5, labels=[1,2,3,4,5])
```

```python
customer_df["RFM_score"] = (
    customer_df["R_score"].astype(str) +
    customer_df["F_score"].astype(str) +
    customer_df["M_score"].astype(str)
)
```

```python
customer_df["RFM_score"].value_counts().head(10)
```

```python
customer_df.to_csv("rfm_results.csv")
```
